# list of stopwords

# Removing stopwords

```python
#empty list to store clean words:
clean_words = []
```

```python
#final frequencing distribution
fdist = FreqDist(clean_words)
fdist.most_common(10)
```

# stemming (porter's stemmer- algorithms)

```python
#import stemming library:
from nltk.stem import PorterStemmer
```

```python
porter = PorterStemmer()
```

```python
word_list = ["Study", "Studying", "Studies", "Studied"]
for w in word_list:
    print(porter.stem(w))
```

```python
from nltk.stem import SnowballStemmer
snowball = SnowballStemmer("english")
```

```python
word_list = ["decreases", "leaves", "Studies", "plays"]
for w in word_list:
    print(snowball.stem(w))
```

# TEXT PREPROCESSING BASIC

# preprocessing fundamentals

#Conversion of input string to lowercase

```python
input_string = "Hello 102 ? Students90 ., This is 50 < YOUR First > PRACTICAL = _ SessION 2001 () on $  786 ' TEXT MinINg"
result1 = input_string.lower()
print("Input String in Lowercase: ")
print("================================")
print(result1)
```

#removal of punctuations

```python
import string
result2 = result1.translate(str.maketrans('','',string.punctuation))
print(result2)
```

```python
#using loops
replace = ""
punctuation = string.punctuation
for char in result1:
    if char not in punctuation:
        replace = replace + char
print("================================")
print(replace)
```

#remove spaces

```python
result3 = result2.strip()
print("================================")
print(result3)

demo = "                 SJCC      "
demo1 = demo.strip()
print(demo)
print(demo1)
```

removing numbers

```python
import re
result4 = re.sub(r'\d+', '', result3)
print("================================")
print(result4)
```

```python
#import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords
sw = set(stopwords.words('english'))
print("================================")
print(sw)
print("================================")
print(len(sw))
print("================================")
```

```python
from nltk.tokenize import word_tokenize
print("Input Text")
print("================================")
print(result4)
print("================================")
print("Output Tokens")
print("================================")
result4_tokens = word_tokenize(result4)
print(result4_tokens)
```

```python
result5 = []
for w in result4_tokens:
    if w not in sw:
        result5.append(w)
print(result5)
```
