```python
# sklearn feature is a module used to extract features in text or image format
#it convert a collection of text doc to matrix of token counts
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
sentences = ["Hello how are you",
             "Hi students are you all good", 
             "Okay lets study bag of words"]
```

```python
sentences
```

```python
cv = CountVectorizer()
```

```python
B_O_W = cv.fit_transform(sentences).toarray()
```
